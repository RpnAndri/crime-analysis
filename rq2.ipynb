```python
# Get the data csv
import pandas as pd

df = pd.read_csv("./data/main.csv")
df = df.replace(",", "", regex=True)
print(df)
```

```python
# Urban + Rural
df["poverty_urban"] = pd.to_numeric(df["poverty_urban"])
df["poverty_rural"] = pd.to_numeric(df["poverty_rural"])
df["poverty"] = df[["poverty_urban", "poverty_rural"]].sum(axis=1)

# Employment
df["employment"] = pd.to_numeric(df["employment_rate"])

# Index + Nonindex
df["index_crime"] = pd.to_numeric(df["index_crime"])
df["nonindex_crime"] = pd.to_numeric(df["nonindex_crime"])
df["crime"] = df[["index_crime", "nonindex_crime"]].sum(axis=1)

# Dataframe
year_reg_pov_emp_crime_df = df[["year", "region", "poverty", "employment", "crime"]]

```

```python
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Group by region
for region, group in df.groupby("region"):
    # Sort by year and interpolate
    group = group.sort_values("year").interpolate()

    # Z-Score standardization
    cols = ["poverty", "employment", "crime"]
    group_z = group.copy()
    group_z[cols] = group[cols].apply(zscore)

    plt.figure(figsize=(10, 6))
    for col in cols:
        plt.plot(group_z["year"], group_z[col], marker="o", label=col.capitalize())
    
    plt.title(f"Region: {region} Time Series")
    plt.xlabel("Year")
    plt.ylabel("Z-Score")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.show()
```

```python
import numpy as np

# Bar Chart (Regional Comparison)

region_avg = df.groupby("region")[["poverty", "employment", "crime"]].mean().sort_values("region").reset_index()

region_list = [
    "ncr",
    "car",
    "1",
    "2",
    "3",
    "4a",
    "4b",
    "5",
    "6",
    "7",
    "8",
    "9",
    "10",
    "11",
    "12",
    "13",
    "barmm",
]
# Sort according to region list
region_avg["region"] = pd.Categorical(region_avg["region"], categories=region_list, ordered=True)
region_avg = region_avg.sort_values("region").reset_index(drop=True)

print(region_avg)


regions = 17
x = np.arange(len(region_avg["region"]))

# Poverty Comparison
plt.figure(figsize=(10, 6))
plt.bar(range(regions), region_avg["poverty"])
plt.title("Average Poverty by Region")
plt.xticks(x, region_avg["region"], rotation=45)
plt.show()

# Employment Comparison
plt.bar(range(regions), region_avg["employment"])
plt.title("Average Employment Rate by Region")
plt.xticks(x, region_avg["region"], rotation=45)
plt.show()

# Crime Comparison
plt.bar(range(regions), region_avg["crime"])
plt.title("Average Crime by Region")
plt.xticks(x, region_avg["region"], rotation=45)
plt.show()
```
